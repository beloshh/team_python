```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# Read the posts column, set the id as the index and remove some unnecessary columns
posts = pd.read_json("json_data/lucid_table_posts.json")
posts.set_index('id',inplace=True)
posts.drop(['status_id','action','post_id'], axis = 1, inplace=True)
posts.head(20)
```

```python
# Read notifications table, drop columns that have null values and the id column
notifications = pd.read_json("json_data/lucid_table_notifications.json")
notifications.dropna(axis = 1,inplace=True)
notifications.drop(['id'], axis = 1, inplace=True)
notifications.head()
```

```python
# Read thoughts table
thoughts = pd.read_json("json_data/lucid_table_thoughts.json")
thoughts.head()
```

```python
# Merge the posts and notifications column on common column user_id
# Drop unnecessary columns
df2 = pd.merge(posts,notifications,on='user_id')
df2.drop(['tags','created_at_x','updated_at_x','created_at_y','updated_at_y'], axis = 1, inplace = True)
df2.head()
```

```python
# Import TFID word vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Create a TFID object with english stop words
tfidf = TfidfVectorizer(stop_words='english')
```

```python
# fill NaN with empty string
posts['content'] = posts['content'].fillna('')
```

```python
# Run the TFID vectorizer on the column of posts
tfidf_matrix = tfidf.fit_transform(posts['content'])
```

```python
tfidf_matrix.shape
```

```python
# carry out a cosine similarity to see how similar posts are
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
indices = pd.Series(posts.index, index = posts['title']).drop_duplicates()
```

```python
# Function to get Recommendations based on input
def get_recommendations(title, cosine_sim = cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse = True)
    sim_scores = sim_scores[1:11]
    content_indices = [i[0] for i in sim_scores]
    print('Articles you might be interested in:\n')
    return set(posts['title'].iloc[content_indices])
```

```python
get_recommendations('Task Number 7')
```
