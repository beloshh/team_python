```python
#import libraries
import pandas as pd
import sqlalchemy
```

```python
#create engine
engine = sqlalchemy.create_engine('mysql+pymysql://root:@localhost:3306/Lucid')
```

```python
#read users table
users = pd.read_sql_table("users", engine, index_col='id')
```

```python
#drop irrelevant columns
users.drop(['image','password','remember_token'],axis = 1, inplace=True)
```

```python
#views users head
users.head()
```

```python
#read users post
posts = pd.read_sql_table('posts', engine,index_col='id')
```

```python
#drop irrelevant columns
posts.drop(['status_id','action','post_id'], axis = 1, inplace=True)
```

```python
#view posts head
posts.head()
```

```python
#read contact table
contact= pd.read_sql_table('contact_settings',engine,index_col='id')
```

```python
#drop NaN
contact.dropna(axis = 1,inplace=True)
```

```python
#view head
contact.head()
```

```python
#read following table
following = pd.read_sql_table('following',engine)
```

```python
following.head()
```

```python
#read notifications table
notifications = pd.read_sql_table('notifications',engine,index_col='id')
```

```python
#drop NaN
notifications.dropna(axis = 1,inplace=True)
```

```python
notifications.head()
```

```python
#read thoughts table
thoughts = pd.read_sql_table('thoughts',engine,index_col='id')
```

```python
thoughts.head()
```

```python
#merging posts and contact table
df1 = pd.merge(posts,contact,on='user_id')
df1.head()
```

```python
#merging posts and notifications
df2 = pd.merge(posts,notifications,on='user_id')
df2.head()
```

```python
#merging posts and thoughts
df3 = pd.merge(posts,thoughts,on='user_id')
df3.head()
```

```python
#merge users and contact
df4 = pd.merge(users,contact, on='email')
df4.head()
```

```python
df2.groupby('title')['action'].count().sort_values(ascending = False).head()
```
