```python
def status(text):
    if text == 'Commented':
        return 1.0
    elif text == 'Followed':
        return 2.0
    elif text == 'Like':
        return 3.0
    elif text == 'Replied':
        return 4.0
    elif text == 'Love':
        return 5.0
```

```python
df2['rating'] = df2['action'].apply(status)
```

```python
df2.head()
```

```python
#create a pivot table
df = df2.pivot_table(index = 'sender_id', columns = 'title', values = 'status', aggfunc='mean')
```

```python
df.head()
```

```python

x = df['CEO -- Story so Far']
```

```python
#finding the correlation of other articles with CEO--story so far article
corr = df.corrwith(x)
```

```python
corr
```

Intuitively, we can see that CEO-- story so far is correlated with the titles with 0.522233 correlation values

```python
#correlating the columns using pearson formula
matrix_corr = df.corr(method='pearson')
```

```python
matrix_corr.head()
```

```python
#posts in which the sender 1 acted on by either liking, loving, commenting and so on
df.iloc[0].dropna()
```

```python
#creating a series for sender corr
sender_corr = pd.Series()

sender_id = 1

for post in df.iloc[sender_id-1].dropna().index:
    corr_list = matrix_corr[post].dropna()*df.iloc[sender_id-1][post]
    sender_corr = sender_corr.append(corr_list)
```

```python
#grouping user correlations by the index and summing up
sender_corr = sender_corr.groupby(sender_corr.index).sum()
```

```python
#create an empty list for posts
post_list = []

for i in range(len(df.iloc[sender_id-1].dropna().index)):
    if df.iloc[sender_id-1].dropna().index[i] in sender_corr:
        post_list.append(df.iloc[sender_id-1].dropna().index[i])
    else:
        pass
sender_corr = sender_corr.drop(post_list)
```

```python
print('Hi, based on the post that you reacted to: \n')
for i in df.iloc[sender_id-1].dropna().index:
    print(i)
print('\n I would suggest you read these posts \n')
for i in sender_corr.sort_values(ascending= False).index[:5]:
    print(i)
```
